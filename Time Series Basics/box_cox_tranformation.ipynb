```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import boxcox
```

```python
df = pd.read_csv('./datasets/airline_passengers.csv', index_col='Month', parse_dates=True)
```

```python
df.head()
```

```python
df['Passengers'].plot(figsize=(20, 8));
```

```python
df['SqrtPassengers'] = np.sqrt(df['Passengers'])
```

aplastar ligeramente la serie temporal, pero la amplitud del patrol temporal todavia se ve incrementando con el tiempo

```python
df['SqrtPassengers'].plot(figsize=(20, 8));
```

```python
df['LogPassengers'] = np.log(df['Passengers'])
```

```python
la tranformacion de logaritmo se puede aprecia que aplano la serie tempporal haciendola parecer mas uniforme con el tiempo
```

```python
df['LogPassengers'].plot(figsize=(20, 8));
```

```python
data, lam = boxcox(df['Passengers'])
```

```python
lam
```

```python
df['BoxCoxPassengers'] = data
df['BoxCoxPassengers'].plot(figsize=(20, 8));
```
